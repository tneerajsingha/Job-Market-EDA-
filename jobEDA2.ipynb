#### 05

```python
plt.style.use('default')
plt.figure(figsize=[10, 7])

# Extracting salary data based on work mode
rem = df.loc[df['Work Mode'] == 'Remote']['Annual Salary']
site = df.loc[df['Work Mode'] == 'Onsite']['Annual Salary']

# Calculate IQR and bounds for remote salaries
Q1_rem = rem.quantile(0.25)
Q3_rem = rem.quantile(0.75)
IQR_rem = Q3_rem - Q1_rem
lower_bound_rem = Q1_rem - 1.5 * IQR_rem
upper_bound_rem = Q3_rem + 1.5 * IQR_rem

# Calculate IQR and bounds for onsite salaries
Q1_site = site.quantile(0.25)
Q3_site = site.quantile(0.75)
IQR_site = Q3_site - Q1_site
lower_bound_site = Q1_site - 1.5 * IQR_site
upper_bound_site = Q3_site + 1.5 * IQR_site

# Filter out the outliers
rem_filtered = rem[(rem >= lower_bound_rem) & (rem <= upper_bound_rem)]
site_filtered = site[(site >= lower_bound_site) & (site <= upper_bound_site)]

keys = ['Remote', 'On-Site']

# Creating the boxplot without outliers
boxes = plt.boxplot([rem_filtered, site_filtered], labels=keys, patch_artist = True, medianprops = {'linewidth':2})

for box in boxes['boxes']:
    #edge color
    box.set(color = '#4286f4', linewidth = 2)
    # Fill Color
    box.set(facecolor = '#e0e0e0')

plt.title('Salary Level of different Work Mode')
plt.xlabel('Work Mode')
plt.ylabel('Monthly Salary')

plt.savefig('05.png')

plt.show()
```

#### 06

```python
plt.style.use('default')
plt.figure(figsize=[10, 7])

x = df.groupby('City')['Monthly Salary'].max().head(5)
a = x.sort_values(ascending = False).head(5)
df2 = pd.DataFrame(a)

df2 = pd.DataFrame(a).reset_index()
df2.columns = ['City', 'Highest Income']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
bars = plt.bar( 'City', 'Highest Income', data = df2, color = colors)

plt.bar_label(bars)

plt.title('Cities with Highest Salary')
plt.xlabel('City')
plt.ylabel('Monthly Salary')

plt.savefig('06.png')


plt.show()
```
